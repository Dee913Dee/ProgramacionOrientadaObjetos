```python
#Constructor

#el constructor es el encargado de inicializzar un objeto de una clase, asignando los valores iniciales de un nuevo objeto.
#En python, la forma correcta de definir un metodo es #__init__(). 


#El método __init__() se encargar de establecer el primer parámetro especial que se llama self, el cual 
#puede especificar otros parámetros siguiendo las mismas reglas que cualquier otra función.

#Ejemplo del primer esquema; En este caso la clase Automovil es el siguiente:


def __init__(self, color, aceleracion):
    self.color = color
    self.aceleracion = aceleracion
    self.velocidad = 0

#Como se puede ver self, define los parámetros color y aceleracion, velocidad que determinan el estado inicial de un objeto de tipo Automovil.
#self.velocidad haria parte del metodo
#En este caso, para instanciar un objeto de tipo automovil, debemos pasar como argumentos el color y la aceleración 
#como vimos en el ejemplo:

  auto1 = Automovil(morado, 30)
    
```

```python
#Atributos de datos

#los atributos de datos no necesitan ser declarados previamente. Un objeto los crea del mismo modo en que se crean
#las variables en Python, es decir, cuando les asigna un valor por primera vez.

    auto1 = Automovil(morado, 30)
    auto2 = Automovil(azul, 20)
    print(auto1.color)
"respuesta por consola" : morado

     print(auto2.color)
"respuesta por consola" :azul

```

```python
#Métodos

#los métodos son las funciones que se definen dentro de una clase y 
#pueden ser referenciadas por los objetos de dicha clase. Sin embargo, los métodos las funciones acelera() y frena() definen
#un parámetro self.

def acelera(self):
    self.velocidad = self.velocidad + self.aceleracion
    
#No obstante, cuando se usan dichas funciones no se pasa ningún argumento. 
#porque la funcion acelera() está siendo utilizada como un método por los objetos de la clase Automovil,
#de tal manera que cuando un objeto referencia a dicha función, realmente pasa su propia referencia como primer parámetro
#de la función.

#Sabiendo esto, podemos entender, por ejemplo, por qué todos los objetos de tipo Automovil pueden referenciar 
#a los atributos de datos velocidad o color. Son inicializados para cada objeto en el método __init__().
```

```python
#Herencia

#En programacion herencia es conocido como la clase padre y las clases hijas heredan funcionalidades, atributos de esa clase padre 
#pero no son identicas. Aprovecha eso que ya existe y se le añaden nuevas cosas, en otras palabras, la herencia es la capacidad
#de reutilizar una clase extendiendo su funcionalidad. 

#Ahora la clase hija puede añadir nuevos atributos, nuevos métodos o redefinirlos.


class AutoEletrico(Automovil):

    parabrisas = 4

    def __init__(self, color, aceleracion, esta_rodando=False):
        super().__init__(color, aceleracion)
        self.esta_rodando = esta_rodando

    def enciende(self):
        self.esta_rodando = True

    def apagado(self):
        self.esta_rodando = False
        


#Para este ejemplo quise enseñar como nuestra clase inicial que ha sido Automovil, de ella se deriban otras opciones, que si
#bien son similares no son identicas, la clase Autolelectrico hereda de la clase Automovil. 
#y para poder hacer esto el nombre de la clase hija va antes del nombre de la clase padre el cual se indica entre paretesis

#La clase AutomElectrico redefine el atributo de clase parabrisas, cambiando su valor a 4 e implementa dos métodos nuevos: 
#enciende y apagado. 

#la opcion super(). permite re definir el metodo __init__ de la clase hija usando la funcionalidad del método de la clase padre
#Como la clase Automovil e es la que define los atributos color y aceleracion, estos se pasan al constructor de la clase padre y
#se crea el atributo de instancia esta_rodando para objetos de la clase AutoElectrico
```
